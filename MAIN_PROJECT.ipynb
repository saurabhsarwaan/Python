## Data Cleaning

```python
# Removing Null Values
df=df.dropna()
df

# removing duplicates
df=df.drop_duplicates()

# replacing values
df["Gender"]=df["Gender"].replace("Unknown","Others")

df["Quantity"]=df["Quantity"].replace("five",5)

df["UnitPrice"]=df["UnitPrice"].replace("??","NaN")

#how to correct date format
#df["PurchaseDate"]=pd.to_datetime(d)




# checking data types of each column
df.dtypes
```

## Feature Engineering

```python
# create age groups
def age_group(age):
    if age<20:
        return "teenager"
    elif age>=20 and age<40:
        return "young adult"
    elif age>=40 and age<60:
        return "adult"
    else:
        return "senior citizen"
df["Age_Group"]=df["Age"].apply(age_group)
df

# create category sales
Category_sales=df.groupby("ProductCategory")["TotalAmount"].sum().reset_index()
Category_sales

# male vs female sales
Gender_sales=df.groupby("Gender")["TotalAmount"].sum().reset_index()
Gender_sales

#city wise sales
city_sales=df.groupby("City")["TotalAmount"].sum().reset_index()
city_sales
```
